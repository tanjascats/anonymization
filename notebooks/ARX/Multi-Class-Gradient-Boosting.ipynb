# Multi-class Classification of Anonymized 'Adult' Dataset - Gradient Boosting

Results of classification are obtained using Rapid Miner becuase it gave better results compared to scikit-learn implementation.

```python
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
from matplotlib.pyplot import figure
```

```python
k_values = [1, 3, 7, 11, 15, 19, 23, 27, 31, 35, 100]
# read f1 scores obtained using Weka 
filepath = "../../data/ARX/without-contin-hie/f1_scores_Gradient_Boosting_RapidMiner.txt"
with open(filepath) as f:
    lines = f.readlines()
f1_scores_1 = [float(score.rstrip('\n')) for score in lines]
```

```python
# read f1 scores obtained using Weka 
filepath = "../../data/ARX/data/multi-class-edu/f1_scores_Gradient_Boosting_RapidMiner.txt"
with open(filepath) as f:
    lines = f.readlines()
f1_scores_2 = [float(score.rstrip('\n')) for score in lines]
```

```python
figure(figsize=(9.7,6))
line1, = plt.plot(k_values, f1_scores_1, marker='o', label='Only categorical anonymized locally')
line2, = plt.plot(k_values, f1_scores_2, marker='o', label='All anonymized locally')
plt.ylabel('f1 score')
plt.xlabel('k')
plt.title('Gradient Boosting')
plt.legend(handles=[line1, line2])
```
